```python
import pandas as pd

df = pd.read_csv("data.csv")
df = df.iloc[: , 1:] # drop index column
df = df.sort_values(["Province", "WeekEnding"])

# remove any nan values from number columns
for i in range(2,10):
    df[df.columns[i]] = pd.to_numeric(df[df.columns[i]], errors ='coerce').fillna(0).astype('int')
    
# break up date into year, month, day columns
dates = df["WeekEnding"].str.split("/", expand = True)
df["Year"] = dates[0].astype(str)
df["Month"] = dates[1].astype(str)
df["Day"] = dates[2].str.split(" ", expand=True)[0].astype(str)
df.drop(columns=["WeekEnding"], inplace = True)

df.to_csv("cleaned_data.csv") 
```

```python
provinces = df["Province"].unique() # get province values
for province in provinces:
    province_df = df[df["Province"] == province] # filter by province
    
    monthly_df = pd.DataFrame(columns=[
        'Province', 'Year', 'Month', 
        'TotalCases', 'MonthlyCases', 
        'TotalDeaths', 'MonthlyDeaths',
        'TotalHospitalized', 'MonthlyHospitalized',
        'TotalICU', 'MonthlyICU',
    ])
    
    for j in range(3):
        for i in range(12):
            # filter data by month and year
            month_df = province_df[province_df["Month"] == str(i + 1)] 
            month_df = month_df[province_df["Year"] == str(2020 + j)]
            
            # sum weekly numbers to get monthly
            monthlyCases = month_df["WeeklyCases"].sum()
            monthlyDeaths = month_df["WeeklyDeaths"].sum()
            monthlyHospitalized = month_df["WeeklyHospitalized"].sum()
            monthlyICU = month_df["WeeklyICU"].sum()
            
            month_total_df = month_df["TotalCases"].values[-1:]
            if month_total_df.size == 0:
                totalCases = 0,
                totalDeaths = 0,
                totalHospitalized = 0,
                totalICU = 0,
            else:
                totalCases = int(month_df["TotalCases"].values[-1:][0]),
                totalDeaths = month_df["TotalDeaths"].values[-1:][0],
                totalHospitalized = month_df["TotalHospitalized"].values[-1:][0],
                totalICU = month_df["TotalICU"].values[-1:][0],
            
            row = {
                'Province': province,
                'Year': 2020+j,
                'Month': i+1, 
                'TotalCases': totalCases[0],
                'MonthlyCases': monthlyCases,
                'TotalDeaths': totalDeaths[0],
                'MonthlyDeaths': monthlyDeaths,
                'TotalHospitalized': totalHospitalized[0],
                'MonthlyHospitalized': monthlyHospitalized, 
                'TotalICU': totalICU[0],
                'MonthlyICU': monthlyICU, 
            }
            
            monthly_df = pd.concat([monthly_df, pd.DataFrame([row])])
        
    monthly_df.to_csv("{}.csv".format(province))
```
